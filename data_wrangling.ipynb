```python
#load packages
import os
import pandas as pd
```

```python
#loads dataframe
df = pd.read_excel('apartment_data.xlsx')
```

```python
df.head()
```

```python
#creates dataframe with listing only in USD and then removes currenct column
usd_df = df.loc[df['currency'] == 'USD']
apt_df = usd_df.drop('currency', axis=1)
```

```python
#creates dataframe with only apartments for rent then removes category column
apartment_df = apt_df.loc[df['category'] == 'housing/rent/apartment']
apt_rent_df= apartment_df.drop('category', axis=1)
```

```python
#drops columns not needed
clean_apt_df= apt_rent_df.drop(['has_photo','amenities','fee','price_display','price_type'], axis=1)
```

```python
#makes time column human readable
clean_apt_df['time'] = pd.to_datetime(clean_apt_df['time'], unit='s')
```

```python
#removes duplicate entries from the dataset
clean_apt_df = clean_apt_df.drop_duplicates()
```

```python
clean_apt_df = clean_apt_df.drop('source', axis=1)
```

```python
#saves as a new file
clean_apt_df.to_excel('clean_apt_data.xlsx', index=False)
```
