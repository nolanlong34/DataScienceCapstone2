```python
#load dependicies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data frame
apt_df = pd.read_excel('clean_apt_data.xlsx')
```

```python
apt_df.describe()
```

```python
#creates dataframe for only california apartments and shows summary statistics of the data
cal_df = apt_df[apt_df['state'] == 'CA']
cal_df.describe()
```

```python
#creates a dataframe for appartments for bay area cities
bay_area_cities = ['San Francisco', 'Oakland', 'San Jose', 'Palo Alto', 'Mountain View', 'Berkeley','Redwood City', 'San Mateo', 'Sunnyvale', 'Santa Clara', 'Menlo Park', 'Cupertino',
                    'San Leandro', 'Hayward', 'Fremont', 'Milpitas', 'Union City', 'Newark', 'San Ramon', 'Pleasanton', 'Dublin', 'Walnut Creek', 'Concord', 'Pittsburg', 'Antioch', 'Brentwood',
                      'Richmond', 'El Cerrito', 'San Pablo', 'Albany', 'El Sobrante', 'Pinole', 'Hercules', 'Rodeo', 'Martinez', 'Pacheco', 'Clayton', 'Pleasant Hill', 'Lafayette', 'Orinda',
                        'Moraga', 'Danville', 'Alamo', 'Blackhawk', 'Diablo', 'Castro Valley', 'San Lorenzo', 'Alameda', 'Emeryville', 'Albany', 'El Cerrito', 'San Pablo', 'El Sobrante', 'Pinole',
                          'Hercules', 'Rodeo', 'Martinez', 'Pacheco', 'Clayton', 'Pleasant Hill', 'Lafayette', 'Orinda', 'Moraga', 'Danville', 'Alamo', 'Blackhawk', 'Diablo', 'Castro Valley',
                            'San Lorenzo', 'Alameda', 'Emeryville', 'Albany', 'El Cerrito', 'San Pablo', 'El Sobrante', 'Pinole', 'Hercules', 'Rodeo', 'Martinez', 'Pacheco', 'Clayton', 
                            'Pleasant Hill', 'Lafayette', 'Orinda', 'Moraga', 'Danville', 'Alamo', 'Blackhawk', 'Diablo', 'Castro Valley', 'San Lorenzo', 'Alameda', 'Emeryville']
bay_df = cal_df[cal_df['cityname'].isin(bay_area_cities)]

```

```python
#describes the data for the bay area
bay_df.describe()
```

```python
#creates a histogram of the price of apartments under 5000 dollars filtier out luxury oultiers above $8000
plt.hist(bay_df['price'][bay_df['price'] < 8000], bins = 50)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Price of Apartments in the Bay Area')
```

```python
#creates a histogram of bedrooms in the bay area
plt.hist(bay_df['bedrooms'], bins = 5)
plt.xlabel('Bedrooms')
plt.ylabel('Count')
plt.title('Number of Bedrooms in the Bay Area')
```

```python
#create a histogram of bathrooms in the bay area
plt.hist(bay_df['bathrooms'], bins = 10)
plt.xlabel('Bathrooms')
plt.ylabel('Count')
plt.title('Number of Bathrooms in the Bay Area')
```

```python
#creates a histogram of the square footage of apartments in the bay area
plt.hist(bay_df['square_feet'], bins = 50)
plt.xlabel('Square Feet')
plt.ylabel('Count')
plt.title('Square Footage of Apartments in the Bay Area')
```

```python
#creates a new column that assigns 0 to NaN for pets_allowed column, 1 for cats, 2 for dogs, and 3 for both
bay_df['pets_allowed'] = bay_df['pets_allowed'].fillna(0)
bay_df['pets_allowed'] = bay_df['pets_allowed'].replace('Cats', 1)
bay_df['pets_allowed'] = bay_df['pets_allowed'].replace('Dogs', 2)
bay_df['pets_allowed'] = bay_df['pets_allowed'].replace('Cats,Dogs', 3)
```

```python
#creates a histogram of the pets_allowed column
plt.hist(bay_df['pets_allowed'], bins = 4)
plt.xlabel('Pets Allowed')
plt.ylabel('Count')
plt.title('Pets Allowed in the Bay Area')
```

```python
#creates a new column for price per square foot which may be a valuable feature
bay_df['price_per_sqft'] = bay_df['price'] / bay_df['square_feet']
```

```python
bay_df
```

```python
#drops unneccessary columns
bay_df = bay_df.drop(columns = ['state', 'id', 'time','title','body'])
```

```python
#saves the dataframe to a new excel file
bay_df.to_excel('bay_area_apt_data.xlsx')
```

```python
#creates a histogram of the price per square foot
plt.hist(bay_df['price_per_sqft'], bins = 50)
plt.xlabel('Price Per Square Foot')
plt.ylabel('Count')
plt.title('Price Per Square Foot in the Bay Area')
```
